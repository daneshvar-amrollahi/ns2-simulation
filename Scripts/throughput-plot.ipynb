```python
import sys
import matplotlib.pyplot as plt
```

```python
file_path = input("Enter file path:")
node = input("Enter node name for throghput calculation:")

#Example: file path:55mbps/wireless.tr
#         node: 8
```

```python
file = open(file_path, 'r')
lines = file.readlines()

total_time = int(float(lines[-1].split()[1]))
receiver_node = "_" + node + "_"

y = [0] * 101

data = 0
for line in lines:
    words = line.split()    
    if words[0] == "r" and words[2] == receiver_node:
        cur_time = int(float(words[1]))
        packet_size = int(words[7])
        y[cur_time] += packet_size
        
x = [i for i in range(1, 102)]

fig, ax = plt.subplots()
ax.plot(x, y, linewidth=2.0)
plt.xlabel("Time (s)")
plt.ylabel("Data received (bits)")

plt.show()
```
